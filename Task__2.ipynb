```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_cols=['Age','Fare','SibSp','Parch']
```

```python
df[numeric_cols].hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle('Histograms of Numeric Features')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[numeric_cols + ['Survived']], hue='Survived', diag_kind='hist')
plt.suptitle("Pairplot with Survival Hue", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_cols + ['Survived']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival Count by Sex")
plt.show()
print(" Inference: Females had a significantly higher survival rate than males.")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival Count by Passenger Class")
plt.show()
print(" Inference: 1st class passengers had the highest survival rate, followed by 2nd, then 3rd.")
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Age', hue='Survived', kde=False, bins=30)
plt.title("Age Distribution by Survival")
plt.show()
print(" Inference: Children under 10 had higher survival. Elderly passengers had lower survival.")
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Fare', hue='Survived', bins=40, kde=False)
plt.title("Fare Distribution by Survival")
plt.show()
print(" Inference: Passengers who paid higher fares were more likely to survive (linked to class).")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='SibSp', hue='Survived')
plt.title("Survival Count by Number of Siblings/Spouses")
plt.show()
print(" Inference: Having 12 siblings/spouses helped survival; too many decreased it.")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Parch', hue='Survived')
plt.title("Survival Count by Number of Parents/Children")
plt.show()
print(" Inference: Having 13 parents/children aboard slightly improved survival chance.")
```
